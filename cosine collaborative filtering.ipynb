```python
import pandas as pd
import numpy as np
from math import sqrt
import matplotlib.pyplot as plt
import collections
import xlsxwriter 
```

```python
#import data set
products_df = pd.read_csv('part_bdml_data.csv')
products_df.head(10)
```

```python
#Split the values
products_df['products'] = products_df.products.str.split('|')
```

```python
x=products_df['products'] 
x
```

```python
#Checking to convert frequency as ratings
p=pd.value_counts(x[0])
p
```

```python
#Checking index
y=p.index
y    
```

```python
#Make rating matrics
a = (80,300)
b=np.zeros(a)
```

```python
#Saving Customer ID
z=products_df['customerId'] 
z
```

```python
#Make customer ID and ratings of products

for i in range(0,80):
    p=pd.value_counts(x[i])
    y=p.index
    for k in range(0,80):
        b[k,0]=z[k]
        for j in range(0,y.size):
            b[i,(int(y[j])+1)]=p[j]
        
print(b)        
    
b=b.T  
```

```python
#Delete customer ID 
inputs=np.delete(b,0,0)
```

```python
#Find the similarity matrics and make diagonal as 0
from sklearn.metrics.pairwise import cosine_similarity
output = cosine_similarity(inputs)
np.fill_diagonal(output, 0 )
ratings=np.dot(output,inputs)
```

```python
#Make a for loop for all the product ID
l=list(range(0,299))
```

```python
#Make Weighted matrics
def user(x):
    m=ratings[:,x]
    o=np.vstack([l,m])
    o=o.T
    v=o[o[:,1].argsort()[::-1]]
    print(type(v))
    return v[:,0]
```

```python
# Find final answer for user 7
answer_final=list((user(7)))
type(answer_final[0])
```

```python
#Previous Product ID 
q=(x[7])
list_set = set(q) 
q = (list(list_set))
q
```

```python
#Mamke product ID to integeer
answer_final = [int(i) for i in answer_final] 
answer_final = list(map(int, answer_final))
answer_final
```
